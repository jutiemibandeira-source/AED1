```python
pessoas = int(input("Digite a quantidade de pessoas"))
soma_peso = 0
soma_altura= 0
imc_maior= 0 
imc_menor= 1000
x = 1
while x <= pessoas:
    peso = float(input("Qual seu peso? "))
    altura = float(input("Qual sua sua altura? "))
    imc = peso / (altura*altura)
    if imc > imc_maior:
        imc_maior = imc
    elif imc < imc_menor:
        imc_menor = imc
    
    soma_altura += altura
    soma_peso += peso
    x +=1
print(soma_altura/pessoas)
print(soma_peso/pessoas)
print(imc_maior)
print(imc_menor)
```
