```python
## Exercício 1
n = int(input("Digite um número"))
if n % 3 == 0 and n % 5 == 0:
    print("O número é divisível")
else: 
    print("O número não é divisível")
```

```python
## Exercício 2
preco = float(input("Qual o valor da sua compra?"))
if preco <= 1000:
    desconto = preco * 0.1
    print(f"O seu desconto  é de {desconto} e você pagará {preco - desconto}")
elif preco >= 1001 and preco <= 5000:
    desconto = preco * 0.2
    print(f"O seu desconto  é de {desconto} e você pagará {preco - desconto}")
else:
    desconto = preco * 0.3
    print(f"O seu desconto  é de {desconto} e você pagará {preco - desconto}")
    
```

```python
## Exercício 03
largura = float(input("Digite a largura do apto: "))
comprimento = float(input("Digite o comprimento do apto: "))
litros = float(input("Digite quantos litros vem em cada lata: "))
altura = 2.8
area_paredes = (2*largura*altura)+(2*altura*comprimento)
area_porta = 0.8 * 2.1
area_total = area_paredes - area_porta
litros_totais = area_total / 3
latas = litros_totais / litros
print(f"Litros totais = {litros_totais}")
print(f"Área total a pintar {area_paredes}")
print(f"Quantidades de latas necessárias {latas}")
```
