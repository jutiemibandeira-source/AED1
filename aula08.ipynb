```python
import random
n = random.randint(0,50)
palpite = 0 
tentativa = 0
while n != palpite:
    tentativa += 1
    if tentativa > 7:
        continue 
    escolha = int(input("Escolha um número de 1 a 50"))
    palpite = escolha 
    if palpite == n:
        print("Parabéns você acertou!!") 
        break 
    elif palpite > n:
        print("O número digitado é maior!")
    else:
        print("O número digitado é menor!")
print("Você não pode mais tentar, o seu  nº de tentativas esgotou!")
```
