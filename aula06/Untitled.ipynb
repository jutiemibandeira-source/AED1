```python
nota1 = float(input("Insira sua primeira nota: "))
if nota1 >=7:
    nota1 += nota1 + 0.5
nota2 = float(input("Insira sua segunda nota: "))
if nota2 >=7:
    nota2 += 0.5
media = (nota1 + nota2)/2 
print(f"Sua média é igual a {media}")
if media >=5:
    print("Aprovado")
else:
    print("Reprovado")
```

```python
a = float(input("Insira o primeiro lado do triângulo"))
b = float(input("insira o segundo lado do triângulo"))
c = float(input("insira o terceiro lado do triângulo"))
if (a + b > c) and (a + c > b) and (b + c > a):
    if a == c == a == b:  
        print("Seu triângulo é equilátero")
    elif a != b != c:
        print("Seu triângulo é escaleno")
    else:
        print("Seu triângulo é isósceles")
else:
    print("Suas medidas não podem formar um triângulo!!")
```
